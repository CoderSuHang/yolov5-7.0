```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 打开视频文件
video = cv2.VideoCapture("./gesture.mp4")
# 读取一帧
ret, frame = video.read()
# video.read()来读取一帧图片
# 产生两个返回值：
# ret(返回值ture or false，即是否读取到数据)
# frame(读取到的内容)
```

```python
# 打印图片
plt.imshow(frame)
# 图片格式是BGR通道
```

```python
# 转换读取图片颜色通道
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
video = cv2.VideoCapture("./gesture.mp4")
num = 0             # 计数器
save_step = 30      # 间隔帧
while True:
    ret, frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./images/" + str(num) + ".jpg", frame)
```
